```python
import pandas as pd 
import utilities
from preprocessing import Preprocessor
```

```python
df = utilities.load_processed()
```

```python
proc = Preprocessor()
```

```python
wordcount = [len(proc.tokenise(tweet)) for tweet in  df['processed_tweet']]
wordcount.sort()
```

```python
import numpy as np
np.array([wordcount])[:50000].std()
```

```python
import matplotlib.pyplot as plt
plt.hist(np.array([wordcount])[:50000], bins=250)
plt.xlabel('Word Count')
plt.ylabel('Tweet Count')
plt.title('Word Count for Processed Tweets')
```
